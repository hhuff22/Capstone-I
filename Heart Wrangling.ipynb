```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
heart_prev = pd.read_csv("heart_disease_restaurants.csv")
heart_prev.drop("Unnamed: 0", axis=1, inplace=True)
heart_prev.head()
```

```python
plt.scatter("Restaurant Count","Weighted Mean Prevalence", data=heart_prev)
plt.xlabel("Restaurant Count")
plt.ylabel("Heart Disease Prevalence")
plt.show()
```

```python
sns.lmplot("Restaurant Count","Weighted Mean Prevalence",data=heart_prev, hue="Population Group",fit_reg=False)
plt.show()
```

```python
popgroups = heart_prev.groupby("Population Group")["Weighted Mean Prevalence"].mean().reset_index()
popgroups
```

```python
restgroups = heart_prev.groupby("Restaurant Group")["Weighted Mean Prevalence"].mean().reset_index()
restgroups
```
